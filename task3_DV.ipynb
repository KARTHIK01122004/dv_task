```python
# ===============================
# Task-3 : Quantity vs Product
# Aim: To visualize bivariate analysis (continuous vs categorical)
# ===============================

# 1. Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from joypy import joyplot
import matplotlib.cm as cm

# Adjust font size for all plots
plt.rcParams.update({'font.size': 8}) # Reduced font size

# 2. Load Data
df = pd.read_csv("sales_data_full.csv")

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# ------------------------------
# 3. Quantity vs Product
# ------------------------------

# (a) Bar Chart: Avg Quantity by Product
plt.figure(figsize=(4,3)) # Reduced figure size
sns.barplot(x="Product", y="Quantity", data=df, estimator="mean", errorbar=None, palette="Set2")
plt.title("Bar Chart: Avg Quantity by Product")
plt.show()

# (b) Kernel Density Plot: Quantity by Product
plt.figure(figsize=(4,3)) # Reduced figure size
sns.kdeplot(data=df, x="Quantity", hue="Product", fill=True, common_norm=False, alpha=0.5, palette="Set2")
plt.title("Kernel Density Plot: Quantity by Product")
plt.show()

# (c) Box Plot: Quantity by Product
plt.figure(figsize=(4,3)) # Reduced figure size
sns.boxplot(x="Product", y="Quantity", data=df, palette="Set2")
plt.title("Box Plot: Quantity by Product")
plt.show()

# (d) Ridgeline Plot: Quantity by Region
plt.figure(figsize=(6,4)) # Reduced figure size
region_order = df.groupby('Region')['Quantity'].mean().sort_values().index
g = sns.FacetGrid(df, row="Region", hue="Region", aspect=4, height=1.5, sharex=True, row_order=region_order) # Adjusted height
g.map(sns.kdeplot, "Quantity", fill=True, alpha=0.6)
g.set_titles("{row_name}")
g.set_xlabels("Quantity")
plt.subplots_adjust(hspace=0.3)
plt.suptitle("Ridgeline Plot: Quantity by Region", y=1.02)
plt.show()

# (e) Beeswarm Plot: Quantity by Product
plt.figure(figsize=(4,3)) # Reduced figure size
product_order = df.groupby('Product')['Quantity'].mean().sort_values().index
sns.swarmplot(x="Product", y="Quantity", data=df, palette="Set2", order=product_order)
plt.title("Beeswarm Plot: Quantity by Product")
plt.show()
```
