```python
# Task 6: Force-based Layout Visualization for CustomerProduct Network

# Step 1: Install required libraries
# Step 2: Import libraries
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Step 3: Load dataset
file_path = "sales_data_full.csv"  # Replace with the correct path if needed
data = pd.read_csv(file_path)

# Inspect column names
print("Columns in dataset:", data.columns.tolist())

# --- Adjust these column names to match your file ---
customer_col = "Region"
product_col = "Product"
# ----------------------------------------------------

# Step 4: Create edge list between customers and products
edges = list(zip(data[customer_col], data[product_col]))

# Step 5: Create a bipartite graph
G = nx.Graph()
G.add_edges_from(edges)

# Step 6: Compute Force-based layout (FruchtermanReingold algorithm)
pos = nx.spring_layout(G, seed=42)

# Step 7: Visualization
plt.figure(figsize=(10, 8))

# Color nodes by type: customers vs. products
customer_nodes = set(data[customer_col])
product_nodes = set(data[product_col])

nx.draw_networkx_nodes(G, pos, nodelist=customer_nodes, node_color="lightblue", node_size=800, label="Customers")
nx.draw_networkx_nodes(G, pos, nodelist=product_nodes, node_color="lightgreen", node_size=800, label="Products")

nx.draw_networkx_edges(G, pos, edge_color="gray", alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=9, font_weight="bold")

plt.title("CustomerProduct Network (Force-based Layout)", fontsize=14)
plt.legend(scatterpoints=1)
plt.axis("off")
plt.show()
```
