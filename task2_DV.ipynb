```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/sales_data_full.csv")
df['Revenue'] = df['Quantity'] * df['Price']

# Set style
sns.set(style="whitegrid")


# Bar Chart - Frequency of each Region
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Region', palette='pastel')
plt.title("Bar Chart: Frequency of Regions")
plt.show()

# Pie Chart - Distribution of Products
product_counts = df['Product'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(product_counts, labels=product_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart: Product Distribution")
plt.show()



# Histogram - Quantity Distribution
plt.figure(figsize=(6,4))
sns.histplot(df['Quantity'], kde=False, color='skyblue')
plt.title("Histogram: Quantity")
plt.show()

# Density Plot - Price Distribution
plt.figure(figsize=(6,4))
sns.kdeplot(df['Price'], shade=True, color='orange')
plt.title("Density Plot: Price")
plt.show()

# Rug Plot - Price
plt.figure(figsize=(6,4))
sns.rugplot(df['Price'], height=0.05)
plt.title("Rug Plot: Price")
plt.show()

# Strip Plot - Quantity per Product
plt.figure(figsize=(6,4))
sns.stripplot(x="Product", y="Quantity", data=df, color='green')
plt.title("Strip Plot: Quantity per Product")
plt.show()

# Swarm Plot - Price per Region
plt.figure(figsize=(8,4))
sns.swarmplot(x='Region', y='Price', data=df)
plt.title("Swarm Plot: Price per Region")
plt.show()

# Scatter Plot - Price vs Quantity
plt.figure(figsize=(6,4))
sns.scatterplot(x='Price', y='Quantity', data=df, color='purple')
plt.title("Scatter Plot: Price vs Quantity")
plt.show()

# Line Plot - Revenue over Dates
df['Date'] = pd.to_datetime(df['Date'])
df_grouped = df.groupby('Date')['Revenue'].sum().reset_index()
plt.figure(figsize=(8,4))
sns.lineplot(data=df_grouped, x='Date', y='Revenue', marker='o')
plt.title("Line Plot: Daily Revenue Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
